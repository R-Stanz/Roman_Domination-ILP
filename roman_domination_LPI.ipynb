```python
import matplotlib.pyplot as plt
import networkx as nx
from ortools.linear_solver import pywraplp
import pandas as pd
from handlers import *
```

```python
cities_names = [
                "Constantinople",
                "Antioch",
                "Ravenna",
                "Ephesus",
                "Carthage",
                "Alexandria",
                "Amorium",
                "Athens",
                "Milan",
                "Thessalonika",
                "Mystras"
                ]
```

```python
G = nx.petersen_graph()
number_of_nodes = len(G.nodes())

pos = nx.shell_layout(G)
```

```python
labels = {n:cities_names[n] for n in range(number_of_nodes)}

nx.draw_networkx_nodes(G, pos, node_size=1000, alpha=0.9)
nx.draw_networkx_edges(G, pos)
x = nx.draw_networkx_labels(G, pos, labels, font_size=15)
```

```python
labels = {n:generate_index_label(n) for n in range(number_of_nodes)}

nx.draw_networkx_nodes(G, pos, node_size=1000, alpha=0.95)
nx.draw_networkx_edges(G, pos)
x = nx.draw_networkx_labels(G, pos, labels, font_size=15)
```
